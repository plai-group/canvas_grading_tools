```python
# custom function to read list of student ids from a file
def get_ids(path):
    return (pd.read_csv(
    path, 
    header=None, 
    names=['cwl'])
            .cwl.str.split("_").str[1]
            .astype(int).to_list())
```

```python
## question 1 quality
ids = get_ids("./grades/q21_no_docstring.txt")
question = '_529'
scores.loc[ids, question] = "B+"
comments.loc[ids, question] = "No doc string"
print(f"Updated {len(ids)} students")
```

```python
## question 22 not done

ids = get_ids("./grades/q22_not_done.txt")
question = '_6577'

scores.loc[ids, question] = 0
comments.loc[ids, question] = "Not complete"
print(f"Updated {len(ids)} students")
```

```python
## question 31 everyone full grades except a few manual updates
```

```python
## question 32 too wide
ids = get_ids("./grades/q32_too_wide.txt")
question = '_2902'
scores.loc[ids, question] = "B"
comments.loc[ids, question] = "Image should be 400x400"
print(f"Updated {len(ids)} students")
```

```python
## question 33 optional: most not done, update those who did
all_students = scores.index
completed_the_question = get_ids("./grades/q33_complete.txt")
question = '_3002'

scores.loc[all_students, question] = 0
scores.loc[completed_the_question, question] = "A+"

print(f"Updated {len(completed_the_question)} students")
```

```python
## question 34 optional: most not done, update those who did
all_students = scores.index
completed_the_question = get_ids("./grades/q34_complete.txt")
question = '_6671'

scores.loc[all_students, question] = 0
scores.loc[completed_the_question, question] = "A+"
print(f"Updated {len(completed_the_question)} students")
```
